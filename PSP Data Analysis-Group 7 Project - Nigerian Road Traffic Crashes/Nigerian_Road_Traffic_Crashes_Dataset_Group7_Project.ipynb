```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing Excel File
```

```python
file_path = '/content/Nigerian Road traffic dataset.xlsx'
data = pd.read_excel(file_path, sheet_name='Cleaned Dataset')
```

```python
data.groupby('Fatalities')['Total_Vehicles_Involved'].mean()
```

```python
#Plotting a barchart of total Crashes by state

plt.figure(figsize= (10,6))
sns.barplot(data = data, x='State', y='Total_Crashes', palette='viridis', estimator = 'mean')
plt.xticks(rotation=90)
plt.title('Total Crashes by state')
plt.show()
```

```python
#Plotting a Line plot of total crashes by year and year_quarter

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year and Quarter', y='Total_Crashes', hue='State', data=data, marker='o')
plt.title('Total Crashes by year and year_quarter')
plt.xlabel('Year and Quarter')
plt.ylabel('Total Crashes')
plt.xticks(rotation=45)
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Plotting a bar chart to show the number of injured by state

plt.figure(figsize= (10,6))
sns.barplot(data = data, x='State', y='Injured', palette='viridis', estimator = 'mean')
plt.xticks(rotation=90)
plt.title('No of Injured per state')
plt.show()
```

```python
#Plotting a bar chart to show Poor weather by state

plt.figure(figsize= (10,4))
sns.barplot(data = data, x='State', y='Poorweather', palette='viridis', estimator = 'mean')
plt.xticks(rotation=90)
plt.title('Poor weather by state')
plt.show()
```

```python
#Plotting a Scatter Matrix / Pairplot to show relationship of key features

from pandas.plotting import scatter_matrix

numeric_columns = ['Total_Crashes',	'Injured',	'Fatalities',	  'Total_Vehicles_Involved',   'Speed Violation',	 'Drunk and driving',  'Poorweather' ]

plt.figure(figsize=(12, 12))
scatter_matrix(data[numeric_columns], alpha=0.8, figsize=(12, 12), diagonal='hist', color='blue')
plt.suptitle('Scatter Matrix of key Features')
plt.show()
```

```python
#Plotting a correlation Heatmap

numeric_data = data.select_dtypes(include=[float, int])

plt.figure(figsize=(10, 8))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
